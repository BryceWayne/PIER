```python
data['Time'] = 60*pd.to_datetime(data['Date'], infer_datetime_format=True).dt.hour + pd.to_datetime(data['Date'], infer_datetime_format=True).dt.minute
data['Depth'] = data['Depth'].astype(float)
data['Temp'] = data['Temp'].astype(float)
data
```

```python
tags = data['Tag'].unique().tolist()
tags
```

```python
def tod(row):
    if row['Time'] < 6*60 or row['Time'] > 18*60:
        return 'Night'
    else:
        return 'Day'
    
data['Time Of Day'] = data.apply (lambda row: tod(row), axis=1)
# data['Location'] = 'S. California'
data.head()
```

```python
depth = data.groupby('Time')['Depth'].mean().tolist()
print(np.mean(depth[0:200]+depth[1200:-1]))
print(np.mean(depth[600:800]))
START, END = 250, 475
LENGTH = END-START
plt.figure(figsize=(10,6))
plt.plot(depth, color='blue', label='Mean Depth')
plt.plot(np.linspace(START, END, LENGTH), depth[START:END], color='red', label='Night to Day')
START += 12*60
END += 12*60
plt.plot(np.linspace(START, END, LENGTH), depth[START:END], color='green', label='Day to Night')
plt.gca().invert_yaxis()
plt.legend()
plt.xlabel('Time')
plt.ylabel('Depth ($m$)')
plt.title('Mean Depth ($m$) vs. Time of Day')
plt.show()
```

```python
# data = data_i
L, H = data['Time'].unique().max(), 600
window = 5
data_ = data[data['Depth'] <= H]
l = np.linspace(0, L+1, L//2).astype(int)
h = np.linspace(0, H+1, H//2).astype(int)
X, Y = np.meshgrid(l, h)
Z = np.sin(X**2 + Y**2)
i = 0
while i < len(l):
    j = 0
    df = data_[data_['Time'] < l[i]+window]
    df = df[l[i]-window <= df['Time']]
    while j < len(h)-1:
        df_ = df[df['Depth'] < h[j]+window]
        df_ = df_[h[j]-window <= df_['Depth']]
        count = df_['Depth'].count()
        Z[j, i] = count
        j += 1
    i += 1


labels = ['Midnight', '3A', '6A', '9A', 'Noon', '3P', '6P', '9P', 'Midnight']
ticks = int(data['Time'].max())
TICKS = []
for i in range(ticks):
    if i % int(ticks/8) == 0:
        TICKS.append(i)

plt.close()
plt.figure(1, figsize=(10,6))
plt.xlabel('Time')
plt.ylabel('Depth ($m$)')
plt.title('Depth ($m$) vs. Time: Swordfish')
plt.pcolormesh(X, Y, Z, cmap='jet')
cbar1 = plt.colorbar()
cbar1.set_label("Count")

plt.xticks(TICKS, labels, rotation=90)
plt.gca().invert_yaxis()
plt.savefig('pics/SF7_heatmap.png', bbox_inches = 'tight')
plt.show()
```
