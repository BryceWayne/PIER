```python
data_i['Date_PST'] = data_i['Date_PST'].astype(str)
data_i['PST'] = data_i['PST'].astype(str)
data_f['Date_PST'] = data_f['Date_PST'].astype(str)
data_f['PST'] = data_f['PST'].astype(str)
```

```python
def clean_time(data):
    ## TAKES VERY LONG TIME without format
    TIMES = data['PST'].tolist()
    times = []
    for time in TIMES:
        times.append(60*int(time.split(':')[0]) + int(time.split(':')[1]))

    data['Time'] = times
    return data
data_i = clean_time(data_i)
data_f = clean_time(data_f)
```

```python
def tod(row):
    if row['Time'] < 6*60 or row['Time'] > 18*60:
        return 'Night'
    else:
        return 'Day'
    
data_i['Time Of Day'] = data_i.apply (lambda row: tod(row), axis=1)
data_i['Location'] = 'S. California'
data_i.head()
```

```python
data_f['Time Of Day'] = data_f.apply (lambda row: tod(row), axis=1)
data_f['Location'] = 'Baja California'
data_f.head()
```

```python
data = data_i
L, H = data['Time'].unique().max(), 600
window = 10
data_ = data[data['Depth'] <= H]
l = np.linspace(0, L+1, L//2).astype(int)
h = np.linspace(0, H+1, H//2).astype(int)
X, Y = np.meshgrid(l, h)
Z = np.sin(X**2 + Y**2)
i = 0
while i < len(l):
    j = 0
    df = data_[data_['Time'] < l[i]+window]
    df = df[l[i]-window <= df['Time']]
    while j < len(h)-1:
        df_ = df[df['Depth'] < h[j]+window]
        df_ = df_[h[j]-window <= df_['Depth']]
        count = df_['Depth'].count()
        Z[j, i] = count
        j += 1
    i += 1


labels = ['Midnight', '3A', '6A', '9A', 'Noon', '3P', '6P', '9P', 'Midnight']
ticks = int(data['Time'].max())
TICKS = []
for i in range(ticks):
    if i % int(ticks/8) == 0:
        TICKS.append(i)

plt.close()
plt.figure(1, figsize=(10,6))
plt.xlabel('Time')
plt.ylabel('Depth ($m$)')
plt.title('Depth ($m$) vs. Time: S. California')
plt.pcolormesh(X, Y, Z, cmap='jet')
cbar1 = plt.colorbar()
cbar1.set_label("Count")

plt.xticks(TICKS, labels, rotation=90)
plt.gca().invert_yaxis()
plt.savefig('pics/BET_scal.png', bbox_inches = 'tight')
plt.show()
```

```python
TIME_WIDTH, DEPTH_WIDTH = 15, 25
L, H = data['Time'].max(), 450
data_ = data[data['Depth'] <= H]
l = np.linspace(0, L+1, L//2).astype(int)
h = np.linspace(0, H+1, H//2).astype(int)
X, Y = np.meshgrid(l, h)
Z = np.sin(X**2 + Y**2)
MONTHS = data['Month'].unique().tolist()
MONTHS.sort()
month_dict ={1: 'Jan',
             2: 'Feb',
             3: 'Mar',
             4: 'Apr',
             5: 'May',
             6: 'Jun',
             7: 'Jul',
             8: 'Aug',
             9: 'Sep',
            10: 'Oct',
            11: 'Nov',
            12: 'Dec'}

for month in MONTHS:
    print(month_dict[month])
    data_m = data_[data_['Month'] == int(month)]
    i = 0
    while i < len(l):
        j = 0
        df = data_m[data_m['Time'] < l[i]+TIME_WIDTH]
        df = df[l[i]-TIME_WIDTH <= df['Time']]
        while j < len(h)-1:
            df_ = df[df['Depth'] < h[j]+DEPTH_WIDTH]
            df_ = df_[h[j]-DEPTH_WIDTH <= df_['Depth']]
            count = df_['Depth'].count()
            Z[j, i] = count
            j += 1
        i += 1

    norm = Z.max()
    Z_ = Z/norm

    plt.figure(1, figsize=(10,6))
    plt.xlabel('Time')
    plt.ylabel('Depth ($m$)')
    plt.title(f'{month_dict[month]} Depth ($m$) vs. Time')
    plt.pcolormesh(X, Y, Z_, cmap='jet')
    cbar1 = plt.colorbar()
    cbar1.set_label("Density")

    plt.xticks(TICKS, labels, rotation=90)
    plt.gca().invert_yaxis()
    plt.savefig(f'pics/BigEyeThresherShark_{month_dict[month]}_heatmap_scaled.png', bbox_inches = 'tight')
#     plt.show()

    plt.close()
    plt.figure(1, figsize=(10,6))
    plt.xlabel('Time')
    plt.ylabel('Depth ($m$)')
    plt.title(f'{month_dict[month]} Depth ($m$) vs. Time')
    plt.pcolormesh(X, Y, Z, cmap='jet')
    cbar1 = plt.colorbar()
    cbar1.set_label("Count")

    plt.xticks(TICKS, labels, rotation=90)
    plt.gca().invert_yaxis()
    plt.savefig(f'pics/BigEyeThresherShark_{month_dict[month]}_heatmap_count.png', bbox_inches = 'tight')
#     plt.show()
```
