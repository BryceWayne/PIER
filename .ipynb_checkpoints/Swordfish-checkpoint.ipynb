```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from pprint import pprint
```

```python
plt.style.use('default')
```

```python
data = pd.read_csv('data/All_swordfish_DST_timeseries_BC_export.txt', sep=",").dropna()
data.sort_values(by=['Depth Adj'], inplace=True)
# print(data.columns)
data.head()
```

```python
data.describe()
```

```python
## TAKES VERY LONG TIME without format
data['Date'] = pd.to_datetime(data['PS Date_Time'], format="%m/%d/%Y %H:%M:%S")
DATES = data['Date'].tolist()
dates, times = [], []
for _ in DATES:
    dates.append(_.date())
    times.append(_.time())
t, times = times, []
for time in t:
    times.append(60*time.hour + time.minute)
data['Date'] = pd.to_datetime(dates)
data['Time'] = times
data = data[['Tag', 'Date', 'Time', 'Depth Adj', 'Temp']]
data.columns = ['Tag', 'Date', 'Time', 'Depth', 'Temp']
data.head()
```

```python
data['Time'].unique().max()
```

```python
data['Depth'].max()
```

```python
data['Month'] = data['Date'].dt.month
data.head()
```

```python
L, H = data['Time'].max(), 450
data_ = data[data['Depth'] <= H]
l = np.linspace(0, L+1, L//2).astype(int)
h = np.linspace(0, H+1, H//2).astype(int)
X, Y = np.meshgrid(l, h)
Z = np.sin(X**2 + Y**2)
i = 0
while i < len(l):
    j = 0
    df = data_[data_['Time'] < l[i]+10]
    df = df[l[i]-10 <= df['Time']]
    while j < len(h)-1:
        df_ = df[df['Depth'] < h[j]+10]
        df_ = df_[h[j]-10 <= df_['Depth']]
        count = df_['Depth'].count()
        Z[j, i] = count
        j += 1
    i += 1

norm = Z.max()
Z_ = Z/norm

plt.figure(1, figsize=(10,6))
plt.xlabel('Time')
plt.ylabel('Depth ($m$)')
plt.title('Depth ($m$) vs. Time')
plt.pcolormesh(X, Y, Z_, cmap='jet')
cbar1 = plt.colorbar()
cbar1.set_label("Density")

plt.xticks(TICKS, labels, rotation=90)
plt.gca().invert_yaxis()
plt.savefig('pics/Swordfish_heatmap_scaled.png', bbox_inches = 'tight')
plt.show()

plt.close()
plt.figure(1, figsize=(10,6))
plt.xlabel('Time')
plt.ylabel('Depth ($m$)')
plt.title('Depth ($m$) vs. Time')
plt.pcolormesh(X, Y, Z, cmap='jet')
cbar1 = plt.colorbar()
cbar1.set_label("Count")

plt.xticks(TICKS, labels, rotation=90)
plt.gca().invert_yaxis()
plt.savefig('pics/Swordfish_heatmap_count.png', bbox_inches = 'tight')
plt.show()
```

```python
TIME_WIDTH, DEPTH_WIDTH = 15, 25
L, H = data['Time'].max(), 450
data_ = data[data['Depth'] <= H]
l = np.linspace(0, L+1, L//2).astype(int)
h = np.linspace(0, H+1, H//2).astype(int)
X, Y = np.meshgrid(l, h)
Z = np.sin(X**2 + Y**2)
MONTHS = data['Month'].unique().tolist()
MONTHS.sort()
month_dict ={1: 'Jan',
             2: 'Feb',
             3: 'Mar',
             4: 'Apr',
             5: 'May',
             6: 'Jun',
             7: 'Jul',
             8: 'Aug',
             9: 'Sep',
            10: 'Oct',
            11: 'Nov',
            12: 'Dec'}

for month in MONTHS:
    print(month_dict[month])
    data_m = data_[data_['Month'] == int(month)]
    i = 0
    while i < len(l):
        j = 0
        df = data_m[data_m['Time'] < l[i]+TIME_WIDTH]
        df = df[l[i]-TIME_WIDTH <= df['Time']]
        while j < len(h)-1:
            df_ = df[df['Depth'] < h[j]+DEPTH_WIDTH]
            df_ = df_[h[j]-DEPTH_WIDTH <= df_['Depth']]
            count = df_['Depth'].count()
            Z[j, i] = count
            j += 1
        i += 1

    norm = Z.max()
    Z_ = Z/norm

    plt.figure(1, figsize=(10,6))
    plt.xlabel('Time')
    plt.ylabel('Depth ($m$)')
    plt.title(f'{month_dict[month]} Depth ($m$) vs. Time')
    plt.pcolormesh(X, Y, Z_, cmap='jet')
    cbar1 = plt.colorbar()
    cbar1.set_label("Density")

    plt.xticks(TICKS, labels, rotation=90)
    plt.gca().invert_yaxis()
    plt.savefig(f'pics/Swordfish_{month_dict[month]}_heatmap_scaled.png', bbox_inches = 'tight')
#     plt.show()

    plt.close()
    plt.figure(1, figsize=(10,6))
    plt.xlabel('Time')
    plt.ylabel('Depth ($m$)')
    plt.title(f'{month_dict[month]} Depth ($m$) vs. Time')
    plt.pcolormesh(X, Y, Z, cmap='jet')
    cbar1 = plt.colorbar()
    cbar1.set_label("Count")

    plt.xticks(TICKS, labels, rotation=90)
    plt.gca().invert_yaxis()
    plt.savefig(f'pics/Swordfish_{month_dict[month]}_heatmap_count.png', bbox_inches = 'tight')
#     plt.show()
```
