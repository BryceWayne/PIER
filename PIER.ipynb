```
def get_data(filename):
    data = pd.read_csv(PATH + f'/{filename}.txt', sep="\t", skip_blank_lines=True, header=None) # If no header then header=None
    print(data)
    data =  data[[1, 2]]
    data.dropna(inplace=True)
    data.columns = data.iloc[0]
    data = data.iloc[1:]
    data.head()
    data = data.astype(np.float32)
    data['depth'] = -1*data['depth'].astype(np.float32)
    return data
```

```
data = get_data('sf01')
plt.plot(data['depth'], '.', label='1')
data = get_data('sf02')
plt.plot(data['depth'], '.', label='2')
data = get_data('sf04')
plt.plot(data['depth'], '.', label='4')
# data = get_data('sf06')
# plt.plot(data['depth'], '.', label='6')
# data = get_data('sf07')
# plt.plot(data['depth'], '.', label='7')
plt.legend()
plt.show()
```
